```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
data=pd.read_csv("Instagram.csv", encoding='latin1')
print(data.head())
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.info()
```

```python
home=data["From Home"].sum()
explore=data["From Explore"].sum()
hashtag=data["From Hashtags"].sum()
others=data["From Other"].sum()

labels=['From Home', 'From Explore', 'From Hashtag', 'From Others']
values=[home,explore,hashtag,others]
fig=px.pie(data, values=values, names=labels, title='Impression on Instagram posts from different sources', hole=0.3)
fig.show()
```

```python
text=" ".join(i for i in data.Caption)
stopwords=set(STOPWORDS)
wordcloud=WordCloud(stopwords=stopwords,background_color="grey").generate(text)
plt.style.use('classic')
plt.figure(figsize=(12,10))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
x=np.array(data[['Likes','Saves','Comments','Shares','Profile Visits','Follows']])
y=np.array(data["Impressions"])
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=PassiveAggressiveRegressor()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
features=np.array([[282.0,233.0,4.0,9.0,165.0,54.0]])
model.predict(features)
```
